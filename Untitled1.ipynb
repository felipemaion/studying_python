```python
#!/usr/bin/env python
# coding: utf-8

# In[1]:


import matplotlib.pyplot as plt
get_ipython().magic(u'matplotlib inline')
import numpy as np

```

```python

# In[26]:


salarios_marcos = np.array([100,200,300,400])
salarios_gileno = np.array([50,300,500,450])


```

```python


# In[33]:


plt.plot(salarios_marcos,c='Black',ls='--',marker='s',ms=4,label='Salarios Marcos')
plt.plot(salarios_gileno,c='Red',marker='^',ms=8,label='Salarios Gileno')
plt.legend(loc='lower right')
plt.show()
        



```

```python


# In[34]:


salarios_pedro = np.array([200,150,300,700])


# In[39]:


plt.plot(salarios_marcos,c='Black',ls='--',marker='s',ms=4,label='Salários Marcos')
plt.plot(salarios_gileno,c='Red',marker='^',ms=8,label='Salarios Gileno')
plt.plot(salarios_pedro,c='Blue',marker='o',ms=8,label='Salarios Pedro')
plt.legend(loc='lower right')
plt.show()


# In[ ]:

```
